This is a spotify DataFrame that has information about playlist information

```python
import pandas as pd

spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')
spotify
```

This is a really long data frame, lets see how many different songs and playlists are on it.

```python
tracks = spotify['track_name'].nunique()
tracks
```

```python
playlists = spotify['playlist_name'].nunique()
playlists
```

Lets see how many times my favorite artist Travis Scott appears on this dataframe

```python
travis = spotify[spotify['artist_name'] == 'Travis Scott']
total_tracks = len(travis)
total_tracks
```

How many playlists does travis scott appear on?

```python
travisplaylists = travis['playlist_name'].nunique()
travisplaylists
```

What percentage of playlists contain at least one travis scott song?

```python
(travisplaylists/playlists)*100
```

Now, lets make a list of all of travis scotts songs and we will sort them by album and song title alphabetically

```python
travislist = travis[['album_name','track_name']]
travis_list = travislist.sort_values(['album_name','track_name'], ascending = [True, True]).drop_duplicates()
travis_list
```

Lets see who the top 10 most popular artists are in this dataframe

```python
popular = spotify
popular = popular['artist_name'].value_counts()
popular.head(10)
```

Where does Travis Scott rank compared to the the most popular artists

```python
popular = popular.reset_index()
popular = popular[popular['count'] == total_tracks]
popular
```
