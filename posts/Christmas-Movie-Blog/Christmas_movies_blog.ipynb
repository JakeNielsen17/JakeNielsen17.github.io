### Christmas Movies

```python
import numpy as np
import pandas as pd
holiday_movies = pd.read_csv("https://bcdanl.github.io/data/holiday_movies.csv")
holiday_movies
```

```python
holiday_movie_genres = pd.read_csv("https://bcdanl.github.io/data/holiday_movie_genres.csv")
holiday_movie_genres
```

Lets see what the lowest rated and the highest rated movies are in this dataframe. (movies with less than 100 votes are filtered out)

```python
low = holiday_movies
low = low[low['num_votes'] >= 100]
low = low.sort_values(['average_rating'], ascending=True)
low.head(15)
```

```python
high = holiday_movies
high = high[high['num_votes'] >= 100]
high = high.sort_values(['average_rating'], ascending=False)
high.head(15)
```

Lets see which movies have the most votes (most popular).

```python
votes = holiday_movies
votes = votes.sort_values(['num_votes'], ascending = False)
votes.head(10)
```

Lets join the genre datafrarme in to see which genre is the nmost popular.

```python
joineddf = pd.merge(holiday_movies, holiday_movie_genres, on='tconst', how='inner')
joineddf
```

```python
topgenres = joineddf
topgenres = topgenres[['genres','primary_title']]
topgenres = topgenres['genres'].value_counts().reset_index()
topgenres.head(15)
```
